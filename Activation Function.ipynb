```python
# SIGMOID ACTIVATION FUNCTION

# ANY OUPUT FROM SIGMOID FUNCTION IS IN THE RANGE BETWEEN O AND 1

import math

def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
sigmoid(1)
```

```python
# TANH ACTIVATION FUNCTION

# ANY OUTPUT WHICH CONVERT INTO -1 AND +1

def tanh(x):
  return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-56)
```

```python
# RELU ACTIVATION FUNCTION

# IT IS IN RANGE OF MAXIMUM 0 TO 1
# ANY NEGATIVE VALUE IT GIVES 0 ANY POSITIVE VALUE RETURNS SAME VALUE

def relu(x):
  return max(0,x)
```

```python
relu(-23)
```

```python
# LEAKY RELU ACTIVATION FUNCTION

# NO CHANGES IN POSITIVE VALUE IF YOU GIVE NEGATIVE VALUE IT MULTIPLES WITH O.1

def leaky_relu(x):
  return max(0.1*x,x)
```

```python
  leaky_relu(-12)
```
